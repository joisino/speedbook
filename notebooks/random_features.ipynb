```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = 100  # 入力ベクトルの次元
n = 1000  # 入力サンプル数
x = np.random.randn(n, d) / np.sqrt(d)  # サンプルの生成
y = np.random.randn(n, d) / np.sqrt(d)  # サンプルの生成
gt = np.exp(-np.sum((x - y) ** 2, axis=1) / 2)  # 真のガウスカーネル値

D = 1000  # ランダム特徴量の次元
b = np.random.rand(D) * 2 * np.pi
w = np.random.randn(D, d)
rx = np.cos(x @ w.T + b) * np.sqrt(2 / D)  # ランダム特徴量
ry = np.cos(y @ w.T + b) * np.sqrt(2 / D)  # ランダム特徴量
rf = (rx * ry).sum(axis=1)  # ランダム特徴量による近似
plt.scatter(gt, rf)
```
