```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
olivetti_faces = fetch_olivetti_faces().data
```

```python
U, S, Vh = np.linalg.svd(olivetti_faces)
```

```python
fig = plt.figure(figsize=(24, 6))
for i in range(10):
    ax = fig.add_subplot(2, 5, i + 1)
    ax.imshow(-Vh[i].reshape(64, 64), cmap='gray') # 基底の i 番目
    ax.axis('off')
plt.show()
```

```python
d = 10
i = 10
approx = (Vh[:d].T @ Vh[:d] @ olivetti_faces[i]).reshape(64, 64)
```

```python
plt.imshow(olivetti_faces[i].reshape(64, 64), cmap='gray') # 元画像
plt.axis('off')
```

```python
plt.imshow(approx, cmap='gray') # 近似画像
plt.axis('off')
```
